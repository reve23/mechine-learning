```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error
```

```python
diabetes = datasets.load_diabetes()

print(diabetes.keys())
```

```python
print(diabetes.data)
```

```python
print(diabetes.DESCR)
```

```python
data_x = diabetes.data[:,np.newaxis,2]
print(data_x)
```

```python
data_x_train = data_x[:-30]
data_x_test = data_x[-30:]
data_y_train = diabetes.target[:-30]
data_y_test = diabetes.target[-30:]

model = linear_model.LinearRegression()
model.fit(data_x_train,data_y_train)

y_predict = model.predict(data_x_test)
print("mean squared is" , mean_squared_error(data_y_test,y_predict))
print("weights: ", model.coef_)
print("Intercept: ",model.intercept_)
```
